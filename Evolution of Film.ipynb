```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
imdb_df = pd.read_excel("IMDb_Top_250.xlsx")

# Assuming 'Genre' is the column containing the movie genres, and 'Rating' is the column containing the IMDb ratings
# Replace them with the actual column names in your dataset

# Split genres into separate columns and take the first genre
imdb_df['FirstGenre'] = imdb_df['genre'].str.split(',').str[0]

# Group by 'FirstGenre' and calculate mean rating
grouped_data = imdb_df.groupby('FirstGenre')['rating'].mean().sort_values(ascending=False)

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Ratings by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
imdb_df = pd.read_excel("IMDb_Top_250.xlsx")

# Assuming 'Year' is the column containing the release years, and 'Rating' is the column containing the IMDb ratings
# Replace them with the actual column names in your dataset

# Group by 'Year' and calculate mean rating
grouped_data = imdb_df.groupby('year')['rating'].mean()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='b')
plt.title('Mean IMDb Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean Rating')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
imdb_df = pd.read_excel("IMDb_Top_250.xlsx")

# Assuming 'AgeCertificate' is the column containing the age certificates, and 'Rating' is the column containing IMDb ratings
# Replace them with the actual column names in your dataset

# Group by 'AgeCertificate' and calculate mode rating
mode_rating_by_certificate = imdb_df.groupby('certificate')['rating'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Drop rows with NaN (no mode) if needed
mode_rating_by_certificate = mode_rating_by_certificate.dropna()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(mode_rating_by_certificate, labels=mode_rating_by_certificate.index, autopct='%1.1f%%', startangle=140)
plt.title('Mode IMDb Rating Distribution by Age Certificate')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


```

```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with the actual file path)
imdb_df = pd.read_excel("IMDb_Top_250.xlsx")

# Assuming 'Rating' is the column containing IMDb ratings
# Replace it with the actual column name in your dataset

# Calculate mean, mode, median, variance, and standard deviation
mean_rating = imdb_df['rating'].mean()
mode_rating = imdb_df['rating'].mode().iloc[0] if not imdb_df['rating'].mode().empty else None
median_rating = imdb_df['rating'].median()
variance_rating = imdb_df['rating'].var()
std_dev_rating = imdb_df['rating'].std()

# Display the results
print(f"Mean Rating: {mean_rating}")
print(f"Mode Rating: {mode_rating}")
print(f"Median Rating: {median_rating}")
print(f"Variance of Ratings: {variance_rating}")
print(f"Standard Deviation of Ratings: {std_dev_rating}")
```
